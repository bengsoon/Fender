# Setting up DataBlock and DataLoaders

```python
# Setting up the DataBlock to 

fender = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=102),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = fender.dataloaders(path)
```

```python
doc(DataBlock)
```

```python
# showing examples of validation set ("valid_pct" = 0.2 of total dataset)
dls.valid.show_batch(max_n=6, nrows=1)
```

## Image Transformation

1. The images we obtained are different in sizes and that will be problematic to be fed as a batch into the GPU. So we need to standardize the images, in this case we use 128x128 px.
1. Below are different methods of image transformation:
    - Adding a padding of "zeros" (black) around the images
    - Squishing the images 
    - Random crops of the images and then zooming in

### Add Padding

```python
fender = fender.new(item_tfms=Resize(128, ResizeMethod.Pad, pad_mode="zeros"))
dls = fender.dataloaders(path)
dls.train.show_batch(max_n=6, nrows=1)
```

_Padding not seem to be useful and causes wastage in our computation that results in lower resolution_

### Squishing

```python
fender = fender.new(item_tfms=Resize(128, ResizeMethod.Squish))
dls = fender.dataloaders(path)
dls.train.show_batch(max_n=6, nrows=1)
```

_Not a good option as it is giving unrealistic shape of the guitars which will cause the model to learn unrealistic feature of each guitar._

### Random Resize Crop

- _Randomly resize the image between the range in (min_scale, 1)_
- _It then crops part of the images_
- _kind of like zooming_


```python
fender = fender.new(item_tfms=RandomResizedCrop(128, min_scale=0.02))
dls = fender.dataloaders(path)
dls.train.show_batch(max_n=6, nrows=1, unique=True)
```

```python
fender = fender.new(item_tfms=RandomResizedCrop(128, min_scale=0.1))
dls = fender.dataloaders(path)
dls.train.show_batch(max_n=6, nrows=1, unique=True)
```

```python
fender = fender.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = fender.dataloaders(path)
dls.train.show_batch(max_n=6, nrows=1, unique=True)
```

- _0.3 as min_scale is a good start to allow us to zoom into the pictures without loosing the important features of the guitars_
- _This means we will preserve 30% of the image before cropping it_
- _Good way to give transformation in the images for the model to learn more about the features._

## Train the Model!

### Image Resolution = 128x128px
- We are going to use RandomResizedCrop (min_scale = 0.3) to augment the images

```python
fender = fender.new(item_tfms=RandomResizedCrop(128, min_scale=0.3))
dls = fender.dataloaders(path)
learn = cnn_learner(dls, resnet50, metrics=error_rate)
learn.fine_tune(4)
```

_60% of accuracy, better than average. But we hope to be better_

### Image Resolution = 224 x 224 px

```python
fender = fender.new(item_tfms=RandomResizedCrop(224, min_scale=0.3))
dls = fender.dataloaders(path)
learn = cnn_learner(dls, resnet50, metrics=error_rate)
learn.fine_tune(4)
```

### Image Res = 224px, min_scale = 0.5?

```python
fender = fender.new(item_tfms=RandomResizedCrop(224, min_scale=0.5))
dls = fender.dataloaders(path)
learn = cnn_learner(dls, resnet50, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

## Cleaning

- Sorting the images by their "loss"
- Each image is labeled with four things: prediction, actual (target label), loss, and probability
- The probability here is the confidence level, from zero to one, that the model has assigned to its prediction:

```python
interp.plot_top_losses(5, nrows=5)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
# delete the images marked as "delete" in the cleaner
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
```

```python
# change the "category" for the images marked for a different category in the cleaner
for idx,cat in cleaner.change(): shutil.move(str(cleaner.fns[idx]), path/cat)
```

## Retrain Model

```python
fender = fender.new(item_tfms=RandomResizedCrop(224, min_scale=0.5))
dls = fender.dataloaders(path)
learn = cnn_learner(dls, resnet50, metrics=error_rate)
learn.fine_tune(4)
```
